# Project Phase 1: Data Exploration

## Project Goal
The goal of this project is to analyze health factors to predict diabetes.
- **Classification Task:** Predict whether a patient has diabetes (Diagnosis: 1) or not (0).
- **Clustering Task:** Group patients with similar health profiles to find hidden patterns.

## Dataset Source
The dataset was obtained from Kaggle.
- **Link:** [Diabetes Health Dataset Analysis](https://www.kaggle.com/datasets/rabieelkharoua/diabetes-health-dataset-analysis)


```python
import pandas as pd

# Define the raw dataset URL from your GitHub repository
url = "https://raw.githubusercontent.com/Layanf76/Mining-project-/main/Dataset/diabetes_data%205.csv"

# Load the dataset into a pandas DataFrame
df = pd.read_csv(url)

# Display the total number of records (instances) and columns (attributes)
print(f"Number of instances (records): {df.shape[0]}")
print(f"Number of attributes (features): {df.shape[1]}")

# Define and analyze the target class attribute
class_label = "Diagnosis"
print(f"\nClass Label: {class_label}")
print("Instance count for each label (0 = No Diabetes, 1 = Diabetes):")
print(df[class_label].value_counts())

# Display data types for each feature to understand the dataset structure
print("\nAttributes Data Types:")
print(df.dtypes)

# Print the first 5 rows of the raw dataset as a sample
print("\nRaw Dataset Sample:")
df.head()
```
